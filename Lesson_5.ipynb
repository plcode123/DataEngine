```python
from wordcloud import WordCloud#导入词云库
import pandas as pd  #导入pandas数据分析库
from PIL import Image #导入 Image库，通过从文件加载图像，作用：处理其他图像，或从头创建图像
import matplotlib.pyplot as plt
'''
加载CSV中数据,读取到pandas
'''
dataset = pd.read_csv('./Market_Basket_Optimisation.csv', header = None)
# 将数据存放到transactions中
transactions= []
for i in range(0, dataset.shape[0]):
    temp = []
    for j in range(0, 20):
        if str(dataset.values[i, j]) != 'nan':
            temp.append(str(dataset.values[i, j]))
    transactions.append(str(temp))
#print(transactions)
cut_text = " ".join(transactions)
wc = WordCloud(
    max_words=100,#定义词云的最大显示词数
    width=2000,#定义词云画布的宽度
    height=1200,#定义词云画布
    )
    #使用generate方法，从文本生成词云
wordcloud = wc.generate(cut_text)
    # 导出图片
wordcloud.to_file("wordcloud.jpg")
    # 屏幕显示词云图
plt.imshow(wordcloud)
    #关闭坐标轴，不显示刻度
plt.axis("off")
```
