```python
from wordcloud import WordCloud#导入词云库
import pandas as pd  #导入pandas数据分析库
from PIL import Image #导入 Image库，通过从文件加载图像，作用：处理其他图像，或从头创建图像

'''
加载CSV中数据,读取到pandas
'''
# 读取title 和 genres字段
data = pd.read_csv('./movies.csv')
#合并并且通过“ ” 来分割所有的“genres”字段，电影的“title”字段不用分析
genres = " ".join(data['genres'])
#print(genres)
'''
定义函数，去除文本中无需分析虚词，也称为“停用词”

'''
def remove_stop_words(f):
    stop_words = ['Movie']# “Movie”无实际分析意义
    for stop_word in stop_words:#遍历字符串
        f = f.replace(stop_word, '')#找到“停用词”后，替代为空字符串
    return f
'''
定义函数，生成词云
'''
def create_word_cloud(f):
    print('根据词频，开始生成词云!')
    #去“停用词”
    f = remove_stop_words(f)
    #生成词云
    wc = WordCloud(
        max_words=100,#定义词云的最大显示词数
        width=2000,#定义词云画布的宽度
        height=1200,#定义词云画布
        )
    #使用generate方法，从文本生成词云
    wordcloud = wc.generate(f)
    # 导出图片
    wordcloud.to_file("wordcloud.jpg")
    # 屏幕显示词云图
    plt.imshow(wordcloud)
    #关闭坐标轴，不显示刻度
    plt.axis("off")
    
'''
调用函数，生成词云
'''
create_word_cloud(genres)
```
